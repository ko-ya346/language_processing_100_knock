30. 形態素解析結果の読み込み  
形態素解析結果（neko.txt.mecab）を読み込むプログラムを実装せよ．ただし，各形態素は表層形（surface），基本形（base），品詞（pos），品詞細分類1（pos1）をキーとするマッピング型に格納し，1文を形態素（マッピング型）のリストとして表現せよ．第4章の残りの問題では，ここで作ったプログラムを活用せよ．

```python
import pandas as pd

df = pd.read_csv("neko.txt.mecab", sep="\t", header=None)
tdf = df[1].str.split(",", expand=True)
data = pd.concat([df[0], tdf], axis=1)
data.columns = ["surface", "pos", "pos1", "pos2", "pos3", "base", "app_type", "cunjugation", "proto", "reading"]
data = data.loc[~data.surface.isin(["\u3000", "EOS"])].reset_index(drop=True)

keys = ["surface", "base", "pos", "pos1"]
dic1 = {}

for key in keys:
    dic1[key] = list(data[key].values)

dic2 = dict(zip(data["surface"].tolist(), data[["pos", "pos1", "base"]].values.tolist()))
```

31. 動詞  
動詞の表層形をすべて抽出せよ．

```python
for k, v in dic2.items():
    if v[0] == "動詞":
        print(k)
```

32. 動詞の基本形  
動詞の基本形をすべて抽出せよ．

```python
for k, v in dic2.items():
    if v[0] == "動詞":
        print(v[2])
```

33. 「AのB」  
2つの名詞が「の」で連結されている名詞句を抽出せよ．

```python
nlen = len(dic1["surface"])
for idx in range(nlen):
    if (dic1["surface"][idx]=="の" 
        and (idx!=0 or idx!=nlen-1) 
        and (dic1["pos"][idx-1]=="名詞" and dic1["pos"][idx+1]=="名詞")):
        print(dic1["surface"][idx-1], dic1["surface"][idx+1])
        
```

34. 名詞の連接  
名詞の連接（連続して出現する名詞）を最長一致で抽出せよ．

```python
ans = ""
tmp = ""

nlen = len(dic1["surface"])

for idx in range(nlen):
    if dic1["pos"][idx]=="名詞":
        tmp += dic1["surface"][idx]
    else:
        tmp = ""
    if len(ans) < len(tmp):
        print(tmp)
        ans = tmp
print(ans)
```

37. 「猫」と共起頻度の高い上位10語  
「猫」とよく共起する（共起頻度が高い）10語とその出現頻度をグラフ（例えば棒グラフなど）で表示せよ．

```python
with open("neko.txt.mecab", "rt", encoding="utf-8") as f:
    sentence = f.read().split("EOS\n")
```

```python
from collections import defaultdict

sentence = list(filter(lambda x: x!="", sentence))
cnt = defaultdict(int)

for se in sentence:
    if "猫" not in se:
        continue
    for neko in se.split("\n"):
        if neko=="":
            continue
        tango = neko.replace("\t", ",").split(",")[0]
        if tango != "猫":
            cnt[tango] += 1
        
cnt = sorted(cnt.items(), key=lambda x: x[1], reverse=True)[:10]
labels = [k for k, _ in cnt]
num = [v for _, v in cnt]
plt.bar(labels, num)
plt.show()
```

38. ヒストグラム  
単語の出現頻度のヒストグラムを描け．ただし，横軸は出現頻度を表し，1から単語の出現頻度の最大値までの線形目盛とする．縦軸はx軸で示される出現頻度となった単語の異なり数（種類数）である．

```python
import pandas as pd

df = pd.read_csv("neko.txt.mecab", sep="\t", header=None)
tdf = df[1].str.split(",", expand=True)
data = pd.concat([df[0], tdf], axis=1)
data.columns = ["surface", "pos", "pos1", "pos2", "pos3", "base", "app_type", "cunjugation", "proto", "reading"]
data = data.loc[~data.surface.isin(["\u3000", "EOS"])].reset_index(drop=True)
```

```python
hist = data.surface.value_counts().reset_index().groupby("surface")["index"].nunique().reset_index()
labels = hist.surface.values
cnt = hist["index"]

plt.bar(labels, cnt)
plt.show()
```
