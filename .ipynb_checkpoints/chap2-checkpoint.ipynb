# 16.自然数Nをコマンドライン引数などの手段で受け取り，入力のファイルを行単位でN分割せよ．同様の処理をsplitコマンドで実現せよ．

```python
-(-1000//3)
```

```python
1000//3
```

```python
n = int(input())
if n <= 1:
    print("2以上の数値を指定してください")
else:
    df = pd.read_table("popular-names.txt", header=None)
    nrow = len(df)//n
    for i in range(n):
        if i == n-1:
            df.iloc[nrow*i:, :].to_csv(f"16/ans_{i}.txt", sep="\t",index=False, header=None)
        else:
            df.iloc[nrow*i:nrow*(i+1), :].to_csv(f"16/ans_{i}.txt", sep="\t",index=False, header=None)
```
